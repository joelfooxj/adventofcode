```python
# Start at the smallest and work your way up... 

def findDiffs(joltList): 
    diff1count = 0 
    diff3count = 0 
    currJ = 0
    for i in range(len(joltList)): 
        if joltList[i] - currJ == 1: 
            diff1count += 1
        elif joltList[i] - currJ == 3: 
            diff3count += 1
        else: 
            print("No valid next node")
        currJ = joltList[i]
    return diff1count, diff3count
```

```python
diff1, diff3 = findDiffs(joltList)
print(diff1, diff3)
print(diff1 * diff3)
```

```python
# all paths are valid 
def constructNodes(joltList): 
    jNodes = {} 
    currJ = 0
    for i in range(len(joltList)): 
        jNodes[currJ] = [currJ + i for i in [1,2,3] if (currJ + i) in joltList]
        currJ = joltList[i]
    jNodes[max(joltList)] = [] 
    return jNodes
```

```python
constructNodes(sampleInput)
```

```python
# check each next node and add number of possible paths from that node to termination 
import time 
def findPath(jNum, jNodes): 
    # time.sleep(1)
    # print("Checking " + str(jNum))
    paths = 0
    terminates = False
    if jNum == max(jNodes): 
        # print("Reached " + str(max(jNodes)))
        return True, 1
    for n in jNodes[jNum]:
        nTerm, nPaths = findPath(n, jNodes)
        if nTerm: 
            paths += nPaths 
            terminates = True
    return terminates, paths
```

```python
jNodes = constructNodes(sampleInput)

start = time.perf_counter()
t, p = findPath(0, jNodes)
end = time.perf_counter()
print(end-start)
print(t, p)
```
